This notebook demonstrates the **Bisection Method** for finding the root of the equation:  
$$f(x) = x^3 - x - 2$$


```

# Bisection Method Implementation in Python
# Equation: f(x) = x^3 - x - 2

def f(x):
    return x**3 - x - 2

def bisection(a, b, tol):
    if f(a) * f(b) >= 0:
        print("Bisection method fails.")
        return None
    iteration = 0
    while (b - a) / 2.0 > tol:
        iteration += 1
        midpoint = (a + b) / 2.0
        if f(midpoint) == 0:
            break
        elif f(a) * f(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint
        print(f"Iteration {iteration}: a={a:.6f}, b={b:.6f}, mid={midpoint:.6f}, f(mid)={f(midpoint):.6f}")
    return (a + b) / 2.0

root = bisection(1, 2, 0.0001)
print("\nApproximate root:", root)
```
